```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs #Draw
from rdkit import Chem
from rdkit import Chem
from rdkit.Chem import rdChemReactions
```

```python
class molecule:
    def __init__(self, key):
    
        # 从key分子创建可读写副本 #asy
        self.smile = Chem.RWMol(key)
        # ... 在key的基础上添加原子和键 ...
        self.smile = self.smile.GetMol()
        Chem.SanitizeMol(self.smile)

        self.nummber = 0  
        self.parents = []
        self.reaction = 'set'
        self.layer = -1
        self.energy = 0

    def getsmile(self):
        return self.smile

    def getnummber(self):
        return self.nummber
        
    def setparents(self,a,b):#一个list，里面每一组parents用一个list装
        self.parents=[{'parents':a,'reaction':b}]
    
    def getparents(self):
        return self.parents

    def addparents(self,a):
        self.parents+=a

    def setreaction(self,b):
        self.reaction=b
    
    def getreaction(self):
        return self.reaction

    def setlayer(self,b):
        self.layer=b
    
    def getlayer(self):
        return self.layer  

    def setenergy(self,c):
        self.energy=c
        
    def getenergy(self):
        return self.energy

    def changesmile(self,smile):
        self.smile=Chem.MolFromSmiles(smile)
```
